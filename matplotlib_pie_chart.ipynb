```python
from matplotlib import pyplot as plt
```

```python
values = [83,29,45]
labels = ['Insan','Heyvan','Qus']
plt.pie(values,labels=labels,wedgeprops={'edgecolor':'black'})
plt.title('Canlilar')
plt.show()
```

```python
#For customize the colors
plt.style.use('fivethirtyeight')
values = [83,29,45]
labels = ['Insan','Heyvan','Qus']

colors = ['orange','red','green']

plt.pie(values,labels=labels,colors=colors,wedgeprops={'edgecolor':'blue'})
plt.title('Canlilar')
plt.show()
```

```python
#To explode any slice


plt.style.use('fivethirtyeight')
values = [83,29,45]
labels = ['Insan','Heyvan','Qus']
colors = ['orange','red','green']
explode = [ 0 , 0 , 0.1 ]

plt.pie(values,labels=labels,colors=colors,wedgeprops={'edgecolor':'blue'},explode=explode)
plt.title('Canlilar')
plt.show()
```

```python
# Other arguments of pie chart

# --> shadow=True
# for rotate chart --> startangle = INT

plt.style.use('fivethirtyeight')
values = [83,29,45]
labels = ['Insan','Heyvan','Qus']
colors = ['orange','red','green']
explode = [ 0 , 0 , 0.1 ]

plt.pie(values,labels=labels,colors=colors,wedgeprops={'edgecolor':'blue'},explode=explode,shadow=True,startangle=90)
plt.title('Canlilar')
plt.show()
```

```python
# For adding percentage --> autopct = '%1.1f%%'
'''
if you want to show percentage symbols, just use %% end of the string like above
'''

plt.style.use('fivethirtyeight')
values = [83,29,45]
labels = ['Computer','Phone','TV']
colors = ['orange','red','green']
explode = [ 0 , 0 , 0.1 ]

plt.pie(values,labels=labels,
        wedgeprops={'edgecolor':'blue'},
        explode=explode,
        shadow=True,
        startangle=90
        ,autopct='%1.1f%%')

plt.tight_layout()
plt.title('Products')
plt.show()

# gcosx.js :) 
```
