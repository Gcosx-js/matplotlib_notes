```python
#data = pd.read_csv("data.csv")
plt.style.use("fivethirtyeight")
ages = [10,12,13,55,51,70,110,100,102,130]
bins = [10,40,90,130]

# 10-40 , 40-90 , 90-130 araligindaki yashlarin sayin goster

#Umumi olaraq X yere bolmek istesek bins deyisenine yansidiriq.

#Meselen bins=6 burda 6 yere bolmeye calisacaq deyerleri, eger uygundursa bolecek

plt.hist(ages,bins=bins,edgecolor = 'red')
```
