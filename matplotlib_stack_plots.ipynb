```python
from matplotlib import pyplot as plt
```

```python

plt.style.use('fivethirtyeight')
minutes = [1,2,3,4,5]
pl1 = [1,1,3,4,8]
pl2 = [2,2,5,2,1]
pl3 = [3,3,2,4,1]
lbs = ['Player 1','Player 2','Player 3']
plt.stackplot(minutes,pl1,pl2,pl3,labels=lbs)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')


months = ['January', 'February', 'March', 'April', 'May', 'June']

#  farkl departmann aylk sat verileri
sales_dept1 = [100, 120, 130, 90, 110, 150]
sales_dept2 = [80, 90, 100, 85, 95, 120]
sales_dept3 = [60, 70, 80, 75, 85, 100]


labels = ['Department 1', 'Department 2', 'Department 3']

plt.stackplot(months, sales_dept1, sales_dept2, sales_dept3, labels=labels)


plt.legend(loc='upper left')

plt.title('Monthly Sales by Department')
plt.xlabel('Months')
plt.ylabel('Sales')


plt.tight_layout()
plt.show()
```
