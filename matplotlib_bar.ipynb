# Load data from CSV file - Method 1 (Using CSV module)

```python
from matplotlib import pyplot as plt

import csv
with open("data.csv") as csv_file:
    data = csv.DictReader(csv_file)
    #{'Responder_id': '1', 'LanguagesWorkedWith': 'HTML/CSS;Java;JavaScript;Python'}
    
    from collections import Counter # count languages
    c_langs = Counter()
    
    for row in data:
        c_langs.update(row['LanguagesWorkedWith'].split(';'))
        
langs = []
pop = []

for i in c_langs.most_common(15):
    langs.append(i[0])
    pop.append(i[1])   
    
langs.reverse()
pop.reverse()

plt.style.use('fivethirtyeight')
plt.barh(langs,pop,label='Populyarligi') #--> Horizontal bar chart

plt.legend()
plt.show()
```

# Load data from CSV - Method 2 (Using Pandas)

```python
import pandas as pd
from matplotlib import pyplot as plt
from collections import Counter
data = pd.read_csv('data.csv')
r = data['LanguagesWorkedWith']
c_langs = Counter() 
for re in r:
        c_langs.update(re.split(';'))
        
langs = []
pop = []

for i in c_langs.most_common(15):
    langs.append(i[0])
    pop.append(i[1])  

langs.reverse()
pop.reverse()

plt.style.use('fivethirtyeight')
plt.barh(langs,pop,label='Populyarligi') #--> Horizontal bar chart

plt.legend()
plt.show()
```
