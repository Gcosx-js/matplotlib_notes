```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('data.csv')
data.head(5)
```

```python
ages = data['Age']
py_devs = data['Python']
js_devs = data['JavaScript']
all_devs = data['All_Devs']
```

```python
plt.plot(ages,py_devs,color='orange',label='Python Devs')
plt.plot(ages,js_devs,color='#00CCCC',linestyle='--',label='Javascript devs')
plt.ylabel('Ages of developers')
plt.xlabel('Salary of developers')
range_limit = 54783
plt.legend()
plt.fill_between(ages,py_devs,range_limit,alpha=0.25,
                where = (py_devs > range_limit),
                interpolate=True,color='green')
plt.fill_between(ages,py_devs,range_limit,alpha=0.5,
                where = (py_devs <= range_limit),
                interpolate=True,color='red')
```

```python

plt.plot(ages,all_devs,color='#00CCCC',linestyle='--',label='All devs')
plt.plot(ages,py_devs,color='orange',label='Python Devs')
plt.tight_layout()
plt.ylabel('Ages of developers')
plt.xlabel('Salary of developers')
range_limit = 54783
plt.legend()
plt.fill_between(ages,py_devs,all_devs,alpha=0.25,
                where = (py_devs > all_devs),
                interpolate=True,color='green')
plt.fill_between(ages,py_devs,all_devs,alpha=0.25,
                where = (py_devs <= all_devs),
                interpolate=True,color='red')
```
